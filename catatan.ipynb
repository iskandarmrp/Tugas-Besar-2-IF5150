```python
# Outlier remover
def remove_outlier(data, col):
    # Determine IQR for column
    col_values = data[col].values
    q25, q75 = np.percentile(col_values, 25), np.percentile(col_values, 75)
    print('Quartile 25: {} | Quartile 75: {}'.format(q25, q75))
    iqr = q75 - q25
    print('IQR: {}'.format(iqr))
    print()

    # Determine the cut off (IQR * 1.5)
    cut_off = iqr * 1.5
    lower, upper = q25 - cut_off, q75 + cut_off
    print('Cut Off: {}'.format(cut_off))
    print('Lower bound: {}'.format(lower))
    print('Upper bound: {}'.format(upper))
    print()

    # Examine instances before removal
    print("Number of instances before removal: {}".format(len(data)))

    # Remove outliers
    data = data.drop(data[(data[col] > upper) | (data[col] < lower)].index)

    # Examine instances after removal
    print("Number of instances after removal: {}".format(len(data)))
    print()

    return data
```

```python
import numpy as np

class NaiveBayes:

    def fit(self, X, y):
        n_samples, n_features = X.shape
        self._classes = np.unique(y)
        n_classes = len(self._classes)

        # calculate mean, var, and prior for each class
        self._mean = np.zeros((n_classes, n_features), dtype=np.float64)
        self._var = np.zeros((n_classes, n_features), dtype=np.float64)
        self._priors = np.zeros(n_classes, dtype=np.float64)

        for idx, c in enumerate(self._classes):
            X_c = X[y == c]
            self._mean[idx, :] = X_c.mean(axis=0)
            self._var[idx, :] = X_c.var(axis=0)
            self._priors[idx] = X_c.shape[0] / float(n_samples)
            

    def predict(self, X):
        y_pred = [self._predict(x) for x in X]
        return np.array(y_pred)

    def _predict(self, x):
        posteriors = []

        # calculate posterior probability for each class
        for idx, c in enumerate(self._classes):
            prior = np.log(self._priors[idx])
            posterior = np.sum(np.log(self._pdf(idx, x)))
            posterior = posterior + prior
            posteriors.append(posterior)

        # return class with the highest posterior
        return self._classes[np.argmax(posteriors)]

    def _pdf(self, class_idx, x):
        mean = self._mean[class_idx]
        var = self._var[class_idx]
        numerator = np.exp(-((x - mean) ** 2) / (2 * var))
        denominator = np.sqrt(2 * np.pi * var)
        return numerator / denominator
```

```python
import numpy as np

def sigmoid(x):
    return 1/(1+np.exp(-x))

class LogisticRegression():

    def __init__(self, lr=0.001, n_iters=1000):
        self.lr = lr
        self.n_iters = n_iters
        self.weights = None
        self.bias = None

    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0

        for _ in range(self.n_iters):
            linear_pred = np.dot(X, self.weights) + self.bias
            predictions = sigmoid(linear_pred)

            dw = (1/n_samples) * np.dot(X.T, (predictions - y))
            db = (1/n_samples) * np.sum(predictions-y)

            self.weights = self.weights - self.lr*dw
            self.bias = self.bias - self.lr*db


    def predict(self, X):
        linear_pred = np.dot(X, self.weights) + self.bias
        y_pred = sigmoid(linear_pred)
        class_pred = [0 if y<=0.5 else 1 for y in y_pred]
        return class_pred
```

```python
from sklearn.metrics import accuracy_score, classification_report

class LogisticRegressionManual:
    def __init__(self, learning_rate=0.01, iterations=1000, regularization=0.1):
        self.learning_rate = learning_rate
        self.iterations = iterations
        self.regularization = regularization
        self.weights = None
        self.bias = None

    def sigmoid(self, z):
        return 1 / (1 + np.exp(-z))

    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0

        for _ in range(self.iterations):
            # Linear model
            linear_model = np.dot(X, self.weights) + self.bias
            y_predicted = self.sigmoid(linear_model)

            # Gradients with regularization
            dw = (1 / n_samples) * np.dot(X.T, (y_predicted - y)) + (self.regularization * self.weights)
            db = (1 / n_samples) * np.sum(y_predicted - y)

            # Update parameters
            self.weights -= self.learning_rate * dw
            self.bias -= self.learning_rate * db

    def predict(self, X):
        linear_model = np.dot(X, self.weights) + self.bias
        y_predicted = self.sigmoid(linear_model)
        return [1 if i > 0.5 else 0 for i in y_predicted]

log_reg = LogisticRegressionManual(learning_rate=0.01, iterations=5000, regularization=0.01)
log_reg.fit(X_train, y_train)

y_pred = log_reg.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
import numpy as np
import matplotlib.pyplot as plt

def euclidean_distance(x1, x2):
    return np.sqrt(np.sum((x1-x2)**2))

class KMeans:

    def __init__(self, K=5, max_iters=100, plot_steps=False):
        self.K = K
        self.max_iters = max_iters
        self.plot_steps = plot_steps

        # list of sample indices for each cluster
        self.clusters = [[] for _ in range(self.K)]

        # the centers (mean vector) for each cluster
        self.centroids = []


    def predict(self, X):
        self.X = X
        self.n_samples, self.n_features = X.shape

        # initialize
        random_sample_idxs = np.random.choice(self.n_samples, self.K, replace=False)
        self.centroids = [self.X[idx] for idx in random_sample_idxs]

        # optimize clusters
        for _ in range(self.max_iters):
            # assign samples to closest centroids (create clusters)
            self.clusters = self._create_clusters(self.centroids)

            if self.plot_steps:
                self.plot()

            # calculate new centroids from the clusters
            centroids_old = self.centroids
            self.centroids = self._get_centroids(self.clusters)

            if self._is_converged(centroids_old, self.centroids):
                break

            if self.plot_steps:
                self.plot()

        # classify samples as the index of their clusters
        return self._get_cluster_labels(self.clusters)


    def _get_cluster_labels(self, clusters):
        # each sample will get the label of the cluster it was assigned to
        labels = np.empty(self.n_samples)
        for cluster_idx, cluster in enumerate(clusters):
            for sample_idx in cluster:
                labels[sample_idx] = cluster_idx

        return labels


    def _create_clusters(self, centroids):
        # assign the samples to the closest centroids
        clusters = [[] for _ in range(self.K)]
        for idx, sample in enumerate(self.X):
            centroid_idx = self._closest_centroid(sample, centroids)
            clusters[centroid_idx].append(idx)
        return clusters

    def _closest_centroid(self, sample, centroids):
        # distance of the current sample to each centroid
        distances = [euclidean_distance(sample, point) for point in centroids]
        closest_idx = np.argmin(distances)
        return closest_idx


    def _get_centroids(self, clusters):
        # assign mean value of clusters to centroids
        centroids = np.zeros((self.K, self.n_features))
        for cluster_idx, cluster in enumerate(clusters):
            cluster_mean = np.mean(self.X[cluster], axis=0)
            centroids[cluster_idx] = cluster_mean
        return centroids

    def _is_converged(self, centroids_old, centroids):
        # distances between old and new centroids, for all centroids
        distances = [euclidean_distance(centroids_old[i], centroids[i]) for i in range(self.K)]
        return sum(distances) == 0

    def plot(self):
        fig, ax = plt.subplots(figsize=(12, 8))

        for i, index in enumerate(self.clusters):
            point = self.X[index].T
            ax.scatter(*point)

        for point in self.centroids:
            ax.scatter(*point, marker="x", color="black", linewidth=2)

        plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from scipy.spatial.distance import pdist, squareform

class KMedoidsMixedData:
    def __init__(self, k, numerical_cols, categorical_cols):
        self.k = k
        self.numerical_cols = numerical_cols
        self.categorical_cols = categorical_cols
        
        # Perbarui inisialisasi OneHotEncoder
        self.encoder = OneHotEncoder(sparse_output=False)
        self.scaler = StandardScaler()
    
    def preprocess_data(self, df):
        # Pisahkan kolom numerik dan kategorik
        numeric_data = df[self.numerical_cols].values
        categorical_data = df[self.categorical_cols].values
        
        # Skala data numerik
        scaled_numeric = self.scaler.fit_transform(numeric_data)
        
        # One-hot encode data kategorik
        encoded_categorical = self.encoder.fit_transform(categorical_data)
        
        # Gabungkan kembali
        processed_data = np.hstack([scaled_numeric, encoded_categorical])
        
        return processed_data
    
    def gower_distance(self, processed_data):
        n_samples = processed_data.shape[0]
        n_features = processed_data.shape[1]
        
        distance_matrix = np.zeros((n_samples, n_samples))
        
        for feature in range(n_features):
            # Hitung jarak untuk setiap fitur
            feature_dist = pdist(processed_data[:, feature].reshape(-1, 1), metric='cityblock')
            feature_dist_square = squareform(feature_dist)
            
            # Normalisasi
            feature_dist_square /= np.max(feature_dist_square)
            
            distance_matrix += feature_dist_square
        
        return distance_matrix / n_features
    
    def fit(self, df):
        # Preprocessing data
        processed_data = self.preprocess_data(df)
        
        # Hitung matriks jarak
        distance_matrix = self.gower_distance(processed_data)
        
        # Implementasi K-Medoids (mirip dengan sebelumnya)
        medoid_indices = np.random.choice(len(df), self.k, replace=False)
        
        for _ in range(100):  # iterasi maksimum
            # Assign cluster
            labels = np.argmin(distance_matrix[medoid_indices], axis=0)
            
            # Logika update medoids (disederhanakan)
            changed = False
            for i in range(len(df)):
                if i not in medoid_indices:
                    # Coba tukar medoid
                    pass  # Implementasi lengkap mirip contoh sebelumnya
            
            if not changed:
                break
        
        # Simpan hasil
        self.labels = labels
        self.medoid_indices = medoid_indices
        
        return self

# Contoh penggunaan
xx = pd.DataFrame({
    'umur': [25, 30, 35, 40, 45],
    'pendapatan': [2000, 2500, 3000, 3500, 4000],
    'kota': ['Jakarta', 'Bandung', 'Surabaya', 'Medan', 'Bali'],
    'status': ['Lajang', 'Menikah', 'Lajang', 'Menikah', 'Lajang']
})

# Pilih kolom tertentu
columns_to_select = [
    'lead_time',
    'stays_in_weekend_nights',
    'stays_in_week_nights',
    'adults',
    'children',
    'babies',
    'market_segment',
    'distribution_channel',
    'customer_type',
    'adr',
    'total_of_special_requests']

# Membuat DataFrame baru hanya dengan kolom yang dipilih
df_new = df[columns_to_select].head(5000)

numeric_columns = df_new.select_dtypes(include=['number']).columns.tolist()
categorical_columns = df_new.select_dtypes(include=['object', 'category']).columns.tolist()

# numeric_columns = ['umur', 'pendapatan']
# categorical_columns = ['kota', 'status']

print(numeric_columns)
print(categorical_columns)

# Buat instance K-Medoids untuk data campuran
kmedoids = KMedoidsMixedData(
    k=3,  # jumlah cluster
    numerical_cols=numeric_columns,
    categorical_cols=categorical_columns
)

# Fit model
result = kmedoids.fit(df_new)

# Tambahkan label cluster ke DataFrame
df_new['cluster'] = result.labels

print(df_new)
```

```python
import numpy as np
import pandas as pd
from sklearn.base import BaseEstimator, ClusterMixin
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import silhouette_score
from scipy.spatial.distance import pdist, squareform
import warnings

class AdvancedKMedoidsClustering(BaseEstimator, ClusterMixin):
    """
    Advanced K-Medoids Clustering for Mixed Data Types
    
    Parameters:
    -----------
    k : int, default=3
        Number of clusters
    numerical_cols : list
        List of numerical column names
    categorical_cols : list
        List of categorical column names
    max_iterations : int, default=300
        Maximum number of iterations
    random_state : int, default=42
        Seed for reproducibility
    verbose : bool, default=False
        Print detailed clustering information
    """
    
    def __init__(
        self, 
        k=3, 
        numerical_cols=None, 
        categorical_cols=None,
        max_iterations=300, 
        random_state=42,
        verbose=False
    ):
        # Validasi input
        if numerical_cols is None and categorical_cols is None:
            raise ValueError("At least one type of column must be specified")
        
        self.k = k
        self.numerical_cols = numerical_cols or []
        self.categorical_cols = categorical_cols or []
        self.max_iterations = max_iterations
        self.random_state = random_state
        self.verbose = verbose
        
        # Inisialisasi preprocessing
        self.scaler = StandardScaler()
        self.encoder = OneHotEncoder(
            sparse_output=False, 
            handle_unknown='ignore'
        )
        
    def _preprocess_data(self, df):
        """
        Preprocess mixed data types
        
        Parameters:
        -----------
        df : pandas.DataFrame
            Input dataframe to preprocess
        
        Returns:
        --------
        numpy.ndarray
            Preprocessed data matrix
        """
        # Validasi kolom
        missing_cols = set(self.numerical_cols + self.categorical_cols) - set(df.columns)
        if missing_cols:
            raise ValueError(f"Columns not found in dataframe: {missing_cols}")
        
        # Preprocessing numerik
        numeric_data = df[self.numerical_cols].values
        scaled_numeric = self.scaler.fit_transform(numeric_data)
        
        # Preprocessing kategorik
        categorical_data = df[self.categorical_cols].values
        encoded_categorical = self.encoder.fit_transform(categorical_data)
        
        # Gabungkan
        processed_data = np.hstack([scaled_numeric, encoded_categorical])
        
        return processed_data
    
    def _gower_distance(self, processed_data):
        """
        Hitung matriks jarak Gower untuk data campuran
        
        Parameters:
        -----------
        processed_data : numpy.ndarray
            Matriks data yang sudah diproses
        
        Returns:
        --------
        numpy.ndarray
            Matriks jarak
        """
        n_samples, n_features = processed_data.shape
        distance_matrix = np.zeros((n_samples, n_samples))
        
        for feature in range(n_features):
            # Hitung jarak Manhattan untuk setiap fitur
            feature_dist = pdist(processed_data[:, feature].reshape(-1, 1), metric='cityblock')
            feature_dist_square = squareform(feature_dist)
            
            # Normalisasi
            max_dist = np.max(feature_dist_square)
            if max_dist > 0:
                feature_dist_square /= max_dist
            
            distance_matrix += feature_dist_square
        
        return distance_matrix / n_features
    
    def _kaufman_initial_medoids(self, distance_matrix):
        """
        Pilih medoids awal menggunakan metode Kaufman
        
        Parameters:
        -----------
        distance_matrix : numpy.ndarray
            Matriks jarak
        
        Returns:
        --------
        numpy.ndarray
            Indeks medoids awal
        """
        n_samples = distance_matrix.shape[0]
        
        # Hitung total jarak untuk setiap sampel
        total_distances = np.sum(distance_matrix, axis=1)
        
        # Pilih k medoids pertama dengan total jarak terkecil
        initial_medoids = np.argsort(total_distances)[:self.k]
        
        return initial_medoids
    
    def _compute_cluster_cost(self, distance_matrix, medoid_indices, labels):
        """
        Hitung total biaya/jarak dalam klaster
        
        Parameters:
        -----------
        distance_matrix : numpy.ndarray
            Matriks jarak
        medoid_indices : numpy.ndarray
            Indeks medoids
        labels : numpy.ndarray
            Label klaster
        
        Returns:
        --------
        float
            Total biaya klaster
        """
        total_cost = 0
        for i, label in enumerate(labels):
            total_cost += distance_matrix[medoid_indices[label], i]
        return total_cost
    
    def fit(self, X, y=None):
        """
        Fit model pada data
        
        Parameters:
        -----------
        X : pandas.DataFrame
            Data untuk clustering
        
        Returns:
        --------
        self : object
            Fitted estimator
        """
        # Set random seed
        np.random.seed(self.random_state)
        
        # Preprocessing data
        processed_data = self._preprocess_data(X)
        
        # Hitung matriks jarak
        distance_matrix = self._gower_distance(processed_data)
        
        # Pilih medoids awal
        medoid_indices = self._kaufman_initial_medoids(distance_matrix)
        
        # Iterasi PAM (Partitioning Around Medoids)
        for iteration in range(self.max_iterations):
            # Assign klaster
            labels = np.argmin(distance_matrix[medoid_indices], axis=0)
            
            # Hitung biaya awal
            current_cost = self._compute_cluster_cost(
                distance_matrix, medoid_indices, labels
            )
            
            improved = False
            
            # Coba tukar setiap non-medoid
            for i in range(len(X)):
                if i not in medoid_indices:
                    for j in range(self.k):
                        # Buat konfigurasi medoids baru
                        new_medoid_indices = medoid_indices.copy()
                        new_medoid_indices[j] = i
                        
                        # Hitung label dan biaya baru
                        new_labels = np.argmin(
                            distance_matrix[new_medoid_indices], 
                            axis=0
                        )
                        new_cost = self._compute_cluster_cost(
                            distance_matrix, 
                            new_medoid_indices, 
                            new_labels
                        )
                        
                        # Update jika biaya menurun
                        if new_cost < current_cost:
                            medoid_indices = new_medoid_indices
                            labels = new_labels
                            current_cost = new_cost
                            improved = True
                            break
                    
                    if improved:
                        break
            
            # Cetak informasi debug jika verbose
            if self.verbose:
                print(f"Iteration {iteration + 1}: Cost = {current_cost}")
            
            # Berhenti jika tidak ada perbaikan
            if not improved:
                break
        
        # Simpan hasil
        self.labels_ = labels
        self.medoid_indices_ = medoid_indices
        self.cluster_cost_ = current_cost
        
        # Hitung silhouette score
        try:
            self.silhouette_score_ = silhouette_score(
                processed_data, 
                labels, 
                metric='precomputed', 
                sample_distance_matrix=distance_matrix
            )
        except Exception as e:
            warnings.warn(f"Silhouette score calculation failed: {e}")
            self.silhouette_score_ = None
        
        return self
    
    def predict(self, X):
        """
        Prediksi klaster untuk data baru
        
        Parameters:
        -----------
        X : pandas.DataFrame
            Data baru untuk prediksi
        
        Returns:
        --------
        numpy.ndarray
            Label klaster untuk data baru
        """
        # Preprocessing data
        processed_new_data = self._preprocess_data(X)
        
        # Hitung matriks jarak
        distance_matrix = self._gower_distance(processed_new_data)
        
        # Prediksi label
        return np.argmin(
            distance_matrix[:, self.medoid_indices_], 
            axis=1
        )
    
    def get_cluster_statistics(self, X):
        """
        Dapatkan statistik untuk setiap klaster
        
        Parameters:
        -----------
        X : pandas.DataFrame
            Data asli
        
        Returns:
        --------
        dict
            Statistik untuk setiap klaster
        """
        cluster_stats = {}
        
        for i in range(self.k):
            cluster_mask = self.labels_ == i
            cluster_data = X[cluster_mask]
            
            # Statistik untuk kolom numerik
            numeric_stats = cluster_data[self.numerical_cols].describe()
            
            # Distribusi untuk kolom kategorik
            categorical_stats = {}
            for col in self.categorical_cols:
                categorical_stats[col] = cluster_data[col].value_counts(normalize=True)
            
            cluster_stats[i] = {
                'size': len(cluster_data),
                'numeric_stats': numeric_stats,
                'categorical_stats': categorical_stats,
                'medoid': X.iloc[self.medoid_indices_[i]]
            }
        
        return cluster_stats
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def analyze_hotel_booking_clusters(df_new, numeric_columns, categorical_columns):
    # Inisialisasi K-Medoids
    kmedoids = AdvancedKMedoidsClustering(
        k=4,  # Misal 4 klaster 
        numerical_cols=numeric_columns,
        categorical_cols=categorical_columns,
        verbose=True,
        random_state=42
    )
    
    # Fit model
    kmedoids.fit(df_new)
    
    # Dapatkan statistik klaster
    cluster_stats = kmedoids.get_cluster_statistics(df_new)
    
    # Visualisasi distribusi klaster
    plt.figure(figsize=(12, 6))
    
    # Distribusi ukuran klaster
    plt.subplot(1, 2, 1)
    cluster_sizes = [stats['size'] for stats in cluster_stats.values()]
    plt.pie(cluster_sizes, labels=[f'Cluster {i}' for i in range(len(cluster_sizes))], 
            autopct='%1.1f%%')
    plt.title('Distribusi Ukuran Klaster')
    
    # Boxplot untuk variabel numerik utama
    plt.subplot(1, 2, 2)
    df_new['cluster'] = kmedoids.labels_
    sns.boxplot(x='cluster', y='lead_time', data=df_new)
    plt.title('Distribusi Lead Time antar Klaster')
    
    plt.tight_layout()
    plt.show()
    
    # Cetak statistik detail
    print("\n--- Statistik Klaster Detail ---")
    for cluster, stats in cluster_stats.items():
        print(f"\nKlaster {cluster}:")
        print(f"Ukuran: {stats['size']} ({stats['size']/len(df_new)*100:.2f}%)")
        
        print("\nMedian Numerik:")
        for col in numeric_columns:
            print(f"{col}: {stats['numeric_stats'].loc['50%', col]:.2f}")
        
        print("\nDistribusi Kategorik:")
        for col in categorical_columns:
            top_categories = stats['categorical_stats'][col].head(3)
            print(f"\n{col}:")
            print(top_categories)
    
    # Silhouette score
    print(f"\nSilhouette Score: {kmedoids.silhouette_score_}")
    
    return kmedoids, cluster_stats

# Fungsi untuk visualisasi klaster menggunakan PCA
def visualize_clusters_pca(df_new, labels):
    from sklearn.decomposition import PCA
    from sklearn.preprocessing import StandardScaler
    
    # Preprocessing
    scaler = StandardScaler()
    numeric_data = df_new[numeric_columns]
    scaled_data = scaler.fit_transform(numeric_data)
    
    # Reduksi dimensi dengan PCA
    pca = PCA(n_components=2)
    pca_result = pca.fit_transform(scaled_data)
    
    # Visualisasi
    plt.figure(figsize=(10, 8))
    scatter = plt.scatter(
        pca_result[:, 0], 
        pca_result[:, 1], 
        c=labels, 
        cmap='viridis', 
        alpha=0.7
    )
    plt.colorbar(scatter)
    plt.title('Visualisasi Klaster menggunakan PCA')
    plt.xlabel('Komponen Utama 1')
    plt.ylabel('Komponen Utama 2')
    plt.show()

# Contoh penggunaan
def main(df_new, numeric_columns, categorical_columns):
    # Jalankan klasterisasi
    kmedoids, cluster_stats = analyze_hotel_booking_clusters(
        df_new, 
        numeric_columns, 
        categorical_columns
    )
    
    # Visualisasi klaster dengan PCA
    visualize_clusters_pca(df_new, kmedoids.labels_)
    
    return kmedoids, cluster_stats

# Jalankan analisis
# Catatan: Anda perlu menjalankan ini di lingkungan dengan data yang sesuai
# kmedoids_model, cluster_stats = main(df_new, numeric_columns, categorical_columns)
```
